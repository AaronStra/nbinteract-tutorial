```python
%matplotlib notebook
from ipywidgets import *
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np


B,K,d = 2,1,1
t = np.linspace(0,10,100)
P = B/(K*np.exp(-B*t)+d)
fig = plt.figure(dpi=100)
ax = fig.add_subplot(1, 1, 1)
ax.set_ylim(0,100)
ax.set_ylabel(r"Population", fontsize = 16)
ax.set_xlabel("Time", fontsize = 16)
ax.grid()
line, = ax.plot(t,P)

def update(B,K,d):
    P = B/(K*np.exp(-B*t)+d)
    line.set_ydata(P)


# Defining Sliders

inter = interact(update,
                 B=widgets.FloatSlider(min=0,max=10,step=0.1,value=2),
                 K=widgets.FloatSlider(min=0,max=1,step=0.01,value=1),
                 d=widgets.FloatSlider(min=0,max=10,step=0.1,value=1),
                )
```
